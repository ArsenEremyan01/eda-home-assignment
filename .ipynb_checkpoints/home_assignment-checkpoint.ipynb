## Data Visualization

    To separate users into 3 groups, you need to start by finding the amount of money spent on advertising by each user.

```python
#find user spending
cleaned_data['user_spending'] = (cleaned_data['impression_eCPM'] / 1000) * cleaned_data['impressions_count']
percentiles_by_adt = dict(cleaned_data.groupby('adt')['user_spending'].apply(lambda x: np.percentile(x, [30, 80])))
```

    The function "profitibility_classifier" is used to classify users into 3 categories (low, medium, and high profitability) based on the amount of money a user has spent on advertising. Percentiles (in this case, the 30th and 80th percentiles) of advertising expenses are calculated for each ad type (adt) to determine the user's classification.

```python
cleaned_data
```

```python
def profitibility_classifier(row):
    adt_type = row['adt']
    spent = row['user_spending']
    percentiles = percentiles_by_adt.get(adt_type)
    if spent <= percentiles[0]:
        return 'low'
    elif spent > percentiles[0] and spent <= percentiles[1]:
        return 'middle'
    else:
        return 'high'

cleaned_data['category'] = cleaned_data.apply(profitibility_classifier, axis=1)
```
